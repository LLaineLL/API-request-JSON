# GLEIF LEI request from API

```python
import requests
import pandas as pd
pd.set_option("display.max_columns", 100)

URL = 'https://api.gleif.org/api/v1/lei-records?filter[entity.legalAddress.country]=LV'
r = requests.get(URL) #, proxies=proxyDict)
x = r.json()
```

```python
page = x['meta']
pag = page.get('pagination')
last_page = pag.get('lastPage')
```

```python
ress = pd.DataFrame()
for i in range(1,last_page+1):
    r = requests.get(URL, params={'page[number]':i})
    if r.status_code == 200:
        rez = r.json()
        rezz = pd.json_normalize(rez['data'])
        ress = ress.append(rezz)
```
